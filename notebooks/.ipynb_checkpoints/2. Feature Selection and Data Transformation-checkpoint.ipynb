## 1. Recap

### 1.1 Business Question

>For any given song, can we derive a song's TRUE genre and subgenre - i.e. by inspecting the song's audio attributes derived by Spotify

Often, songs are given a genre subjectively by music managers' opinion. Developing an in-house model for categorising genre based on attributes, as well as a popularity measure, can be of value particularly when tailored for use alongside private data held by the song copyright owners.

**We will implement 2 learning algorithms as follows:**

1. a supervised learning model where we tell the algorithm what genres the songs belong to, based on subjective 'human' genre labels


2. an unsupervised learning model where we ask the algorithm to cluster songs based on their attributes, without seeing the songs' existing genre labels

**We will then see what clusters are derived by the unsupervised learning model and deduce:**

1. can define the song clusters in a way that is understandable to the average human music listener? we can cross reference the cluster outputs from this algorithm against the genre classes from the supervised learning algorithm to deteremine whether the genres are 'natural' genres or not


2. can we attribute song clusters as more or less valuable than others and deduce what song attributes contribute most to song value (based on popularity/ consumption)?

### 1.2 Modeling Expectations

For the supervised learning algorithm we will perform hyperparamter optimisation and model selection to choose a best fit model for multiclass classification. We will build a pipeline and perform grid search, using a Standard Scaler and PCA to transform the data within the pipeline. Grid search will implement a 5 fold cross validation as default, we will optimise for that within the pipeline also.

The models we will look at are:

- KNN
- SVM (multiclass)
- Random Forests

For the unsupervised learning algorithm we will fit and compare 3 clustering methods to see which is the best fit for our dataset:

- K-Means
- DBSCAN
- Hierarchical Clustering

If we have time we will also look at how Neural Networks perform at classifying our songs.

We currently have 2 dataframes, one for the supervised algorithm and one for the unsupervised algorithm - they are the same except for the additional 6000+ rows for the 'alternative' genre in the unsupervised dataset.

## 2. Supervised Dataset Transformations


## DISCUSS DISTRIBUTIONS OF THE FEATURES

### 2.3 Scaling - supervised dataset

We need to scale the data to use in our modelling as we will be using distancing measures in our modelling. Unscaled, some of our columns are in the range 0 to 1 whereas others are in the hundreds of thousands (e.g. duration_ms).

We will scale our data using the minmax scaler.

```python
# instantiate the scaler
supervised_minmaxscaler = MinMaxScaler()

# Transform data
supervised_songs_df_scaled = supervised_minmaxscaler.fit_transform(supervised_songs_df.drop('genre', axis=1))
```

```python
# GET BACK GENRE TAGS
supervised_songs_df_scaled = pd.DataFrame(supervised_songs_df_scaled)
supervised_songs_df_scaled['genre'] = supervised_songs_df['genre']
supervised_songs_df_scaled
```

```python
# PUT BACK COLUMN NAMES
old_colnames = list(supervised_songs_df_scaled.columns)
new_colnames= list(supervised_songs_df.columns)
rename_dict = {a:b for a,b in zip(old_colnames,new_colnames)}
supervised_songs_df_scaled.rename(columns=rename_dict, inplace=True)
supervised_songs_df_scaled
```

### 2.5 Features Profile by Genre

Let's look at the feature profile by genre, plotting the means for each feature by genre in a bar chart.

```python
feature_means_genre = supervised_songs_df_scaled.groupby(by=supervised_songs_df_scaled['genre'], axis=0).mean()
# feature_means_genre.reset_index(inplace=True)
feature_means_genre
```

```python
feature_means_genre.T
```

```python
# create bar chart by attribute for each genre
i=0
for col in feature_means_genre.T.columns:
    plt.title(col)
    plt.barh(feature_means_genre.T.index, feature_means_genre.T.iloc[:,i])
    plt.show()
    i=i+1
```

The genres look very similar to each other in profile - we need to diversify the genres to get genre with more varied profile attributes.

For example, in the above, rap and hiphop conceptually are very similar and often difficult to distinguish, as are rock and alt_rock.

We will go ahead with the steps for the unsupervised dataset and run some clustering to see if we get any clusters, but will have to come back to finding a dataset with more varied songs and genres.

### 2.5 Visualing the data - PCA and t-SNE

```python
# run a PCA with 2 components and plot with genre predictors

# instantiate PCA transformer
pca = PCA(n_components = 2)
supervised_songs_df_PCA = pca.fit_transform(supervised_songs_df_scaled.drop('genre', axis=1))
```

```python
plt.scatter(x = supervised_songs_df_PCA[:, 0], y = supervised_songs_df_PCA[:, 1])
```

```python
pca = PCA(n_components = 7)
supervised_songs_df_PCA = pca.fit_transform(supervised_songs_df_scaled.drop('genre', axis=1))
PCA_df = pd.DataFrame(supervised_songs_df_PCA)
sns.pairplot(PCA_df)
```

```python
PCA_df = pd.DataFrame(pca.components_)
```

```python
# create dataframe to see what components have been given to the X_train features so we can plot
old_colnames = list(PCA_df.columns)
new_colnames= list(supervised_songs_df_scaled.drop('genre', axis=1).columns)
rename_dict = {a:b for a,b in zip(old_colnames,new_colnames)}
PCA_df.rename(columns=rename_dict, inplace=True)
PCA_df
```

```python
# CALCULATE COMPONENT MEANS FOR EACH FEATURE
column_means = []
for c in PCA_df.columns:
    PCA_df[c].mean()
    column_means.append(PCA_df[c].mean()*1000)

PCA_means_df = pd.DataFrame(PCA_df.columns,column_means)
PCA_means_df.reset_index(inplace=True)
PCA_means_df['PCA_means'] = PCA_means_df['index']
PCA_means_df.drop(columns='index', axis=1, inplace=True)
PCA_means_df
```

```python
# let's look only at features that have a mean above 1 or -1
# we will do that by creating a column caled 'keep feature'
PCA_means_df['keep feature'] = PCA_means_df['PCA_means'].abs() >= 1
PCA_means_df=PCA_means_df[PCA_means_df['keep feature'] == True]
PCA_means_df = PCA_means_df.sort_values('PCA_means', ascending=False)
```

```python
# PLOT THE FEATURES ON Y AXIS AND COMPONENT MEANS ON X AXIS
plt.barh(PCA_means_df.iloc[:,0], PCA_means_df['PCA_means'])
```

### 4.3 PCA and t-SNE for Visualisation

To understand whether our clustering is likely to perform well, we want to see if our data is easily separated in groups. It's difficult to see whether we have distinct groups within the above 13x13 pairplots so let's use PCA as a visualisation tool.


```python
pca = PCA(n_components = 4)
pca_data = pca.fit_transform(unsupervised_songs_df_scaled)
sns.pairplot(pd.DataFrame(pca_data))
```

```python
# Instantiate t-SNE. Set verbose = 0 to not print out all the details
tsne = TSNE(n_components = 3, random_state = 4, verbose = 0)
# Fit_Transform t-SNE
tsne_data = tsne.fit_transform(unsupervised_songs_df_scaled)
```

```python
# create tSNE dataframe
tsne_df = pd.DataFrame(tsne_data, columns = [f'tSNE D{i+1}' for i in range(tsne_data.shape[1])])
```

```python
# visualise
sns.pairplot(tsne_df, plot_kws = {'alpha': 0.5})
```

We can see some evidence of groupings/ clusterings so let's get into the clustering mdoels with this dataset and see what the outcome looks like.
