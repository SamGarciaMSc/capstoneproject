## 1. Recap

### 1.1 Business Question

>For any given song, can we derive a song's TRUE genre and subgenre - i.e. by inspecting the song's audio attributes derived by Spotify

Often, songs are given a genre subjectively by music managers' opinion. Developing an in-house model for categorising genre based on attributes, as well as a popularity measure, can be of value particularly when tailored for use alongside private data held by the song copyright owners.

**We will implement 2 learning algorithms as follows:**

1. a supervised learning model where we tell the algorithm what genres the songs belong to, based on subjective 'human' genre labels


2. an unsupervised learning model where we ask the algorithm to cluster songs based on their attributes, without seeing the songs' existing genre labels

**We will then see what clusters are derived by the unsupervised learning model and deduce:**

1. can define the song clusters in a way that is understandable to the average human music listener? we can cross reference the cluster outputs from this algorithm against the genre classes from the supervised learning algorithm to deteremine whether the genres are 'natural' genres or not


2. can we attribute song clusters as more or less valuable than others and deduce what song attributes contribute most to song value (based on popularity/ consumption)?

### 1.2 Modeling Expectations

For the supervised learning algorithm we will perform hyperparamter optimisation and model selection to choose a best fit model for multiclass classification. We will build a pipeline and perform grid search, using a Standard Scaler and PCA to transform the data within the pipeline. Grid search will implement a 5 fold cross validation as default, we will optimise for that within the pipeline also.

The models we will look at are:

- KNN
- SVM (multiclass)
- Random Forests

For the unsupervised learning algorithm we will look at 3 clustering methods to see which is the best fit for our dataset:

- K-Means
- DBSCAN
- Hierarchical Clustering

If we have time we will also look at how Neural Networks perform at classifying our songs.

### 2.2 Address Class Imbalance and Prepare Model Datasets

Let's create 2 or 3 datasets for training and testing our genre classfication models.

The datasets will be subsets containing 3 or 4 conceptually different genres with balanced class representation.

We will train all of the models on each subset and investigate the results.

To start, we will revisit the genre_mappings that were applied to the SQLlite dataset in notebook 4, reducing the number of datapoints in the classes with higher representation in a targeted fashion to optimise the purity of each genre class.

That has been done by removing any worldwide national variations as there could be local musical/folk themes and variations projected onto the music that may make it differ from other regional varieties. To simplify the classes and come up with classes containing datapoints that are as similar as possible, the author has decided to remove all regional varities except american and uk regions (as the author is very familiar with music from those regions and understands the nature of music from those regions). However, Scottish, Irish and Welsh variants have been removed in case of any local/ folk variants.

The following steps rebuilds the tracks_genres_df dataset using the new more conservative mappings to reduce the classes with many more datapoints than others:

```python
# let's start with the joined SQLite dataset exported from notebook 4
# import joined SQL lite dataframe from csv, version before genre mappings were applied
tracks_genres_df_unmapped = pd.read_csv('../data/SQL_tracks_genres_unmapped.csv')
tracks_genres_df_unmapped
```

```python
# drop Unnamed: 0
tracks_genres_df_unmapped.drop('Unnamed: 0', axis=1, inplace=True)
tracks_genres_df_unmapped
```

```python
# import alternative second pass genre mapping table removing regional variants
alt_genres_mapping = pd.read_csv('../data/second_pass_genres_mapping.csv', index_col=0)
alt_genres_mapping
```

```python
# create new_genres_df
new_tracks_genres_df = pd.merge(tracks_genres_df_unmapped, alt_genres_mapping, on='genre_id', how='inner')
new_tracks_genres_df.shape
```

```python
# drop rows highlighted for deletion by the mapping table from new_genres_df
new_tracks_genres_df = new_tracks_genres_df[~(new_tracks_genres_df['genre_group']=='delete')]
new_tracks_genres_df
```

```python
new_tracks_genres_df.drop(['artist_id', 'genre_id'], axis=1, inplace=True)
```

```python
new_tracks_genres_df.duplicated().value_counts()
```

```python
new_tracks_genres_df.drop_duplicates(keep='first', inplace=True)
```

```python
new_tracks_genres_df.shape
```

```python
# find songs that have more than one artist and therefore genre (as in notebook 4)
value_counts = new_tracks_genres_df.id.value_counts()

multi_genre_songs_list = list(value_counts[value_counts > 1].index)

new_tracks_genres_df = new_tracks_genres_df[~new_tracks_genres_df['id'].isin(multi_genre_songs_list)]
new_tracks_genres_df['id'].value_counts()
```

```python
new_tracks_genres_df['genre_group'].value_counts()
```

```python
# combine genres as in notebook 4
new_tracks_genres_df = new_tracks_genres_df.replace(to_replace={'soul':'rnb'})
new_tracks_genres_df = new_tracks_genres_df.replace(to_replace={'reggae':'latino/caribbean', \
                                                                      'ska':'latino/caribbean', 'latin':'latino/caribbean',\
                                                                      'caribbean':'latino/caribbean'})
new_tracks_genres_df = new_tracks_genres_df.replace(to_replace={'disco':'edm', 'dance':'edm', 'electro':'edm'})
new_tracks_genres_df = new_tracks_genres_df.replace(to_replace={'asian pop':'pop'})
new_tracks_genres_df = new_tracks_genres_df.replace(to_replace={'trap':'rap/hiphop'})
```

```python
new_tracks_genres_df['genre_group'].value_counts()
```

We could create datasets of the following class combinations, for classes which are diverse from one another (and also contain fairly balanced class representation as they share similar number of datapoints), and train these 5 datasets separately through our models to see how they do:

- edm, classical, rock
- latino/caribbean, pop, rap/hiphop
- jazz, indie, folk, hardcore_rock
- ballroom, gospel, country
- funk, rnb, blues

Alternatively, we could train our models only on the more core classes that a music fund would invest in, as one multi-class dataset, dropping some datapoints to achieve class balance:

- edm
- classical
- rock
- latino/caribbean
- pop
- rap/hiphop

We'll proceed with the second method to remain aligned to our business question.

We'll drop the higher classes so that there are 9000 datapoints in each class (rap/hiphop will have 6 less but we can live with that).

**N.B. we haven't compared or combined rock and hardcore_rock as the author is adamant that they are very different genres, regardless of whether they share similar attribute profiles. Hardcore_rock is not generally a target for music funds.**

```python
# sample 9000 random rows for each genre
edm_df = new_tracks_genres_df[new_tracks_genres_df['genre_group'] == 'edm'].sample(n=9000)
classical_df = new_tracks_genres_df[new_tracks_genres_df['genre_group'] == 'classical'].sample(n=9000)
rock_df = new_tracks_genres_df[new_tracks_genres_df['genre_group'] == 'rock'].sample(n=9000)
latino_caribbean_df = new_tracks_genres_df[new_tracks_genres_df['genre_group'] == 'latino/caribbean'].sample(n=9000)
pop_df = new_tracks_genres_df[new_tracks_genres_df['genre_group'] == 'pop'].sample(n=9000)
rap_hiphop_df = new_tracks_genres_df[new_tracks_genres_df['genre_group'] == 'rap/hiphop']

# combine into final dataframe
supervised_genres_df_final = pd.concat((edm_df,classical_df, rock_df, latino_caribbean_df, pop_df, rap_hiphop_df), axis=0)
supervised_genres_df_final.reset_index(inplace=True, drop=True)
supervised_genres_df_final
```

```python
supervised_genres_df_final['genre_group'].value_counts()
```

```python
# let's keep the track ids etc in this export in case it's useful at some point
supervised_genres_df_final.to_csv('../data/supervised_genres_df_final.csv')
```

```python
# create version of dataframe with only song features for modeling
# drop IDs, URLs and duration
supervised_genres_df_final.drop(['id','analysis_url', 'duration', 'track_id'], axis=1, inplace=True)
supervised_genres_df_final
```

### 2.3 Train Test and Validation Split
For our supervised learning algorithm we need to do a train test split so that we can set aside some unseen data, to see how our models perform on that unseen data once we have a fine-tuned model that we are happy with.

We will keep aside 25% as our test set, and then 25% of the remain set as our validation set.

```python
# create test set
# set x (features) and y (target) columns
X = supervised_genres_df_final.drop(['genre_group'], axis=1)
y = supervised_genres_df_final['genre_group']

# create train test split keeping 25% for our test set
X_remain, X_test, y_remain, y_test = train_test_split(X,y,test_size=0.25, stratify=y, random_state=1)
```

```python
# create validation and train set
# set x (features) and y (target) columns
X = X_remain
y = y_remain

# create train test split keeping 25% for our test set
X_train, X_val, y_train, y_val = train_test_split(X,y,test_size=0.25, stratify=y, random_state=1)
```

```python
from tempfile import mkdtemp
cachedir = mkdtemp()

# set up estimators and pipeline
estimators = [
    ('scaling', MinMaxScaler()),
    ('dimension_red', PCA()),
    ('classifier', SVC())
]

pipe = Pipeline(estimators, memory = cachedir)

# Create Parameter grid
param_grid = [
    {
        'classifier': [SVC()],
        'scaling': [MinMaxScaler()],
        'dimension_red': [PCA()],
        'dimension_red__n_components': [0.9, 3, 6],
        'classifier__kernel': ['linear', 'rbf'],
        'classifier__C': [1, 10, 100]
#     },
#     {
#         'classifier': [KNN()],
#         'scaling': [StandardScaler(), MinMaxScaler()],
#         'dimension_red':[PCA()],
#         'dimension_red__n_components': [0.9, 3, 6],
#         'classifier__n_neighbors': [2, 5, 10, 20]
#     },
#     {
#         'classifier': [RandomForestClassifier()],
#         'scaling':[None],
#         'dimension_red':[PCA()],
#         'dimension_red__n_components': [0.9, 3, 6],
#         'classifier__n_estimators': [20, 50, 100],
    }
]

# Run GridSearchCV
grid = GridSearchCV(pipe, param_grid, cv = 5)

fitted_grid = grid.fit(X_remain, y_remain)

fitted_grid.best_params_
```
